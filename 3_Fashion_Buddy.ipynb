```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
test=pd.read_csv('fashion-mnist_test.csv')
```

```python
train=pd.read_csv('fashion-mnist_train.csv')
```

```python
train.shape
```

```python
test.shape
```

```python
train_x=train.iloc[:,1:]
train_y=train.iloc[:,0:1]
```

```python
train_x.shape
```

```python
train_y.shape
```

```python
train_x=train_x.values.reshape(-1,28,28)
```

```python
train_y=train_y.values.reshape(-1)
```

```python
train_x.shape
```

```python
train_y.shape
```

```python
test_x=test.iloc[:,1:]
```

```python
test_y=test.iloc[:,0:1]
```

```python
test_x=test_x.values.reshape(-1,28,28)
```

```python
test_y=test_y.values.reshape(-1)
```

```python
test_x.shape
```

```python
test_y.shape
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandle', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
train_x=train_x/255.0
test_x=test_x/255.0
```

```python
train_y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.yticks([])
    plt.xticks([])
    plt.imshow(train_x[i],cmap=plt.cm.binary)
    plt.xlabel(class_names[train_y[i]])
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.losses import SparseCategoricalCrossentropy
```

```python
model=Sequential()
```

```python
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='adam', loss=SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
model.summary()
```

```python
model.fit(train_x, train_y, epochs=10)
```

```python
loss, accuracy = model.evaluate(test_x,test_y)
```

```python
print (loss)
```

```python
print(accuracy*100,"%")
```

```python
predictions= model.predict(test_x)
```

```python
plt.figure(figsize=(20,20))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(test_x[i],cmap=plt.cm.binary)
  if np.argmax(predictions[i]) == test_y[i]:
        color = 'blue'
  else:
        color = 'red'

  plt.xlabel(class_names[np.argmax(predictions[i])]+" \n Actual"+class_names[test_y[i]],color=color)
  print()
```
